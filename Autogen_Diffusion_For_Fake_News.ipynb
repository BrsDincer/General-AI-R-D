````python
class ErrorModule(object):
  def __init__(self)->ClassInitial:
    self.error = NotImplementedError(NotImplemented)
  def __str__(self)->str:
    return "Error Modulation"
  def __call__(self)->ErrorInitial:
    return self.error
  def __getstate__(self)->ErrorInitial:
    raise self.error
  def __repr__(self)->DocumentationInitial:
    return ErrorModule.__doc__
  @property
  def Default(self)->ErrorInitial:
    raise self.error
  def Manuel(self,errorType:ErrorInitial,errorMessage:str)->ErrorInitial:
    raise errorType(errorMessage)
````

````python
class SelectDevice(object):
  def __init__(self)->ClassInitial:
    self.device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
  def __str__(self)->str:
    return "Selection Device Modulation"
  def __call__(self)->NullInitial:
    return None
  def __getstate__(self)->ErrorInitial:
    ErrorModule().Default
  def __repr__(self)->DocumentationInitial:
    return SelectDevice.__doc__
````

````python
class ReadTXT(object):
  def __init__(self,filePath:str)->ClassInitial:
    self.path = filePath
    self.ops = None
  def __str__(self)->str:
    return "Reading TXT Modulation"
  def __call__(self)->NullInitial:
    return None
  def __getstate__(self)->ErrorInitial:
    ErrorModule().Default
  def __repr__(self)->DocumentationInitial:
    return ReadTXT.__doc__
  def Read(self)->ProcessInitial:
    return self.ops.read()
  def __enter__(self)->ProcessInitial:
    self.ops = open(self.path,"r",errors="ignore",encoding="utf-8")
    return self
  def __exit__(self,ex:ClassInitial,eb:ClassInitial,el:ClassInitial)->ProcessInitial:
    self.ops.exit()
````

````python
class ImageProcess(object):
  def __init__(self)->ClassInitial:
    pass
  def __str__(self)->str:
    return "Image Reading Modulation"
  def __call__(self)->NullInitial:
    return None
  def __getstate__(self)->ErrorInitial:
    ErrorModule().Default
  def __repr__(self)->DocumentationInitial:
    return ImageProcess.__doc__
  def ReturnImage(self,image:ImageInitial or str)->ImageInitial:
    img = cv.cvtColor(cv.imread(image),cv.COLOR_BGR2RGB) if isinstance(image,str) else image
    return img
  def ResizeImage(self,image:ImageInitial or str,scaleInitial:tuple=(1000,1000))->ImageInitial:
    img = cv.resize(cv.cvtColor(cv.imread(image),cv.COLOR_BGR2RGB),scaleInitial) if isinstance(image,str) else cv.resize(image,scaleInitial)
    return img
````

# Interested Bounding Box

````python
class BoundingBoxRegionEngine(object):
  def LoadModel(self,modelName:str="yolov8s.pt")->ErrorInitial:
    ErrorModule().Manuel(NotImplementedError,"LoadModel")
  def GetBounding(self,model:ModelInitial,image:ImageInitial,confidence:float or int,inference:int or float,classes:list,save:bool,saveTXT:bool,saveCrop:bool):
    ErrorModule().Manuel(NotImplementedError,"GetBounding")
````

````python
class BoundingBoxRegion(BoundingBoxRegionEngine):
  def __init__(self)->ClassInitial:
    self.device = str(SelectDevice().device)
  def __str__(self)->str:
    return "Bounding Box Detection Modulation"
  def __call__(self)->NullInitial:
    return None
  def __getstate__(self)->ErrorInitial:
    ErrorModule().Default
  def __repr__(self)->DocumentationInitial:
    return BoundingBoxRegion.__doc__
  def LoadModel(self,modelName:str="yolov8s.pt")->ErrorInitial:
    return YOLO(str(modelName))
  def GetBounding(self,model:ModelInitial,image:ImageInitial,confidence:float or int=0.4,inference:int or float=0.7,classes:list=[0],save:bool=False,saveTXT:bool=False,saveCrop:bool=False,isResized:bool=False):
    if isResized:
      img = ImageProcess().ResizeImage(image)
    else:
      img = ImageProcess().ReturnImage(image)
    result = model.predict(
        source=img,
        conf=confidence,
        iou=inference,
        save=save,
        save_crop=saveCrop,
        save_txt=saveTXT,
        classes=classes,
        device=self.device,
        verbose=False,
        show=False,
        visualize=False
    )
    if result is not None:
      points = result[0].boxes.data
      rec = [x[:4].tolist() for x in points] if len(points) > 0 else []
    else:
      rec = []
    return rec
````

# Prediction Mask Visualization

````python
class PredictionMaskOutput(object):
  def __init__(self)->ClassInitial:
    self.color = np.array([30/255,144/255,255/255,0.6])
    self.device = str(SelectDevice().device)
    self.figsize = (8,8)
  def __str__(self)->str:
    return "Prediction Mask Output Modulation"
  def __call__(self)->NullInitial:
    return None
  def __getstate__(self)->ErrorInitial:
    ErrorModule().Default
  def __repr__(self)->DocumentationInitial:
    return PredictionMaskOutput.__doc__
  def GetRandomColor(self)->np.ndarray:
    return np.concatenate(
        [
            np.random.randint(3),
            np.array([0.6])
        ],
        axis=0
    )
  def SigmoidMask(self,mask:ImageInitial,color:np.ndarray)->ImageInitial:
    seg = torch.sigmoid(mask.squeeze(1))
    seg = seg.cpu().numpy().squeeze()
    seg = (seg > 0.5).astype(np.uint8)
    hgt,wdt = seg.shape[-2:]
    mask = seg.reshape(wdt,hgt,1)*color.reshape(1,1,-1)
    return mask
  def Display(self,image:ImageInitial,mask:ImageInitial,isRandom:bool=False,save:bool=False,count:int=0)->ProcessInitial:
    if isRandom:
      color = self.GetRandomColor()
    else:
      color = self.color
    mask = self.SigmoidMask(mask=mask,color=color)
    image = cv.resize(image,(mask.shape[0],mask.shape[1]))
    figure,axs = plt.subplots(figsize=self.figsize)
    axs.axis("off")
    axs.title.set_text("Output-Image Mask")
    axs.imshow(image)
    axs.imshow(mask,alpha=0.6)
    if save:
      if not os.path.exists(os.path.join(os.getcwd(),"predictions")):
        CreateDirectory("predictions")
      path = os.path.join(os.getcwd(),"predictions",f"image_mask_output_{count}.jpg")
      figure.savefig(path)
      plt.tight_layout()
      plt.show()
````

````python
def ManuelMaskPoints(mask:ImageInitial)->list:
  yIDX,xIDX = np.where(mask>0)[:2]
  xMIN,xMAX = np.min(xIDX),np.max(xIDX)
  yMIN,yMAX = np.min(yIDX),np.max(yIDX)
  if mask.shape[-1] == 3:
    h,w,_ = mask.shape
  else:
    h,w = mask.shape
  xMIN = max(0,xMIN-np.random.randint(0,20))
  xMAX = min(w,xMAX+np.random.randint(0,20))
  yMIN = max(0,yMIN-np.random.randint(0,20))
  yMAX = min(h,yMAX+np.random.randint(0,20))
  return [xMIN,yMIN,xMAX,yMAX]
````

````python
def ApplyManuelBox(mask:ImageInitial,palette:np.ndarray)->ImageInitial:
  colorMask = np.zeros((mask.size[0],mask.size[1],3),dtype=np.uint8)
  for z,o in enumerate(palette):
    colorMask[mask==z,:] = o
  return colorMask
````

## Model Structure

````python
class DiffusionStructure(object):
  def __init__(self)->ClassInitial:
    self.size = (512,512)
    self.device = str(SelectDevice().device)
  def __str__(self)->str:
    return "Diffusion Structure Modulation"
  def __call__(self)->NullInitial:
    return None
  def __getstate__(self)->ErrorInitial:
    ErrorModule().Default
  def __repr__(self)->DocumentationInitial:
    return DiffusionStructure.__doc__
  def LoadGenerator(self,seed:int=32)->ProcessInitial:
    return torch.Generator(self.device).manuel_seed(seed)
  def LoadModel(self,isXL:bool=False,modelName:str="runwayml/stable-diffusion-inpainting")->ModelInitial:
    if isXL:
      pipe = StableDiffusionInpaintPipeline.from_pretrained(
          "stabilityai/stable-diffusion-xl-base-1.0",
          torch_dtype=torch.float32,
          cache_dir=os.getcwd(),
          device_map="auto"
      )
    else:
      pipe = StableDiffusionInpaintPipeline.from_pretrained(
          modelName,
          torch_dtype=torch.float32,
          cache_dir=os.getcwd(),
          device_map="auto"
      )
    if self.device.lower() == "cuda":
      pipe = pipe.to(self.device.lower())
    else:
      pass
    return pipe
  def Run(
      self,
      model:ModelInitial,
      image:ImageInitial,
      mask:ImageInitial,
      prompt:str,
      negativePrompt:str,
      height:int=768,
      width:int=768,
      inferenceStep:int=20,
      strength:float=0.3,
      modelName:str="runwayml/stable-diffusion-inpainting",
      isGenerator:bool=False,
      seed:int=5,
      guidance:float=7.5,
      isSaved:bool=True
  )->ProcessInitial:
    if sum(image.size) != sum(mask.size):
      if sum(image.size) > sum(mask.size):
        mask = mask.resize((image.size[0],image.size[1]))
      else:
        image = image.resize((mask.size[0],mask.size[1]))
    if isGenerator:
      generator = self.LoadGenerator(seed=seed)
      result = model(
          prompt=prompt,
          negative_prompt=negativePrompt,
          mask_image=mask,
          image=image,
          height=height,
          width=width,
          num_inference_steps=inferenceStep,
          generator=generator,
          guidance_scale=guidance
      )
    else:
      result = model(
          prompt=prompt,
          negative_prompt=negativePrompt,
          mask_image=mask,
          image=image,
          height=height,
          width=width,
          num_inference_steps=inferenceStep,
          guidance_scale=guidance
      )
    if isSaved:
      ImageSave(os.path.join(os.getcwd(),"diffusion_result.jpg"),result[0][0])
    return result
````

## Diffusion with AE & ControlNET

````python
def LoadDiffusionComplex(aeType:str,netType:str,modelID:str,modelType:ModelInitial)->ModelInitial:
  net = LoadControlNet(netType)
  ae = LoadAutoEncoder(aeType)
  pipe = modelType.from_pretrained(
      modelID,
      controlnet=net,
      vae=ae,
      torch_dtype=torch.float32,
      cache_dir=os.getcwd()
  )
  return pipe
````

````python
def ComplexPrediction(
    pipe:ModelInitial,
    image:ImageInitial,
    prompt:str,
    negativePrompt:str,
    height:int=768,
    width:int=768,
    isGenerator:bool=False,
    inference:int=10,
    guidanceScale:int=10,
    controlScale:int=10,
    seed:int=15,
    controlCondition:float=0.5
)->ProcessInitial:
  if isGenerator:
    generator = torch.Generator(str(SelectDevice().device)).manuel_seed(seed)
  else:
    generator = None
  canny = cv.Canny(image,50,50)
  canny = canny[:,:,None]
  inputImage = np.concatenate([canny,canny,canny],axis=2)
  imagePIL = PImage.fromarray(inputImage)
  DisplayOperation().MultiDisplay(image,inputImage)
  result = pipe(
      prompt=prompt,
      image=imagePIL,
      negative_prompt=negativePrompt,
      height=height,
      width=width,
      guidance_scale=guidanceScale,
      control_scale=controlScale,
      num_inference_steps=inference,
      control_conditioning_scale=controlCondition,
      output_type="array",
      generator=generator
  )
  return result
````

If padding_mask_crop is not None, it will first find a rectangular region with the same aspect ration of the image and contains all masked area, and then expand that area based on padding_mask_crop. The image and mask_image will then be cropped based on the expanded area before resizing to the original image size for inpainting. ## Google Search

````python
def GoogleSearch(topic:Optional[str],similarity:Optional[int]=1,limit:Optional[int]=1)->str:
  wrapper = GoogleSearchAPIWrapper(k=int(similarity),siterestrict=False)
  response = wrapper.results(query=str(topic),num_results=int(limit))
  if (response is not None) and (len(response) != 0):
    if limit > 1:
      content = ""
      for idx in range(len(response)):
        data = response[idx]
        content += (
            f"Title: {data['title']}\n"
            f"Link: {data['link']}\n"
            f"Information: {data['snippet']}"
        )
    else:
      data = response[0]
      content = (
          f"Title: {data['title']}\n"
          f"Link: {data['link']}\n"
          f"Information: {data['snippet']}"
      )
    return content
  else:
    return "[The requested amount of data was not found]"
````

## Wikipedia Search

````python
def WikipediaSearch(topic:Optional[str],limit:Optional[int],**load_kwargs:Any)->list:
  links = []
  contents = []
  response = wpp.search(str(topic),results=int(limit))
  if (len(response) > 0) and (response is not None):
    for idx in response:
      try:
        url = wpp.page(
            str(idx),
            **load_kwargs
        ).url
        links.append(url)
        content = wpp.page(
            str(idx),
            **load_kwargs
        ).content
        contents.append(content)
      except:
        pass
    if (len(links) > 0) and (len(contents) > 0):
      return links,contents
    else:
      return None,None
  else:
    return "[The requested amount of data was not found]"
````

## Arxiv Search

````python
def ArxivSearch(topic:Optional[str])->str:
  wrapper = ArxivAPIWrapper(
      top_k_results=5,
      ARXIV_MAX_QUERY_LENGTH=500,
      load_max_docs=5,
      load_all_available_meta=True,
      doc_content_chars_max=10000
  )
  response = wrapper.run(str(topic))
  if (response is not None):
    return response
  else:
    return "[The requested amount of data was not found]"
````

## News Search

````python
def NewsSearch(urls:Optional[list])->str:
  if isinstance(urls,str):
    urls = [urls]
  loader = NewsURLLoader(
      urls=urls,
      text_mode=True,
      continue_on_failure=True,
      show_progress_bar=True,
      nlp=True
  )
  content = ""
  data = loader.load()
  for idx in range(len(data)):
    main = data[idx]
    try:
      content += data[0].metadata["summary"]
    except:
      content += data[0].metadata["description"]
  if (content is not None) and (content != " "):
    return content
  else:
    return "[The requested amount of data was not found]"
````

## Creating Knowledge Data

````python
def CreateKnowledge(topic:Optional[str],limit:Optional[int]=1)->str:
  googleResponse = GoogleSearch(str(topic),similarity=1,limit=int(limit))
  wikipediaLinks,wikipediaResponse = WikipediaSearch(str(topic),limit=int(limit))
  arxivResponse = ArxivSearch(str(topic))
  content = ""
  if isinstance(wikipediaLinks,list):
    try:
      newsResponse = NewsSearch(wikipediaLinks)
      content += "Knowledge from links:\n"
      content += str(newsResponse)+"\n\n"
    except:
      newsResponse = None
  if isinstance(wikipediaResponse,list):
    content += "Knowledge from wikipedia:\n"
    for element in wikipediaResponse:
      content += str(element)+"\n"
    content += "\n"
  content += "Knowledge from articles:\n"
  content += str(arxivResponse)+"\n\n"
  return content
````

## Result Callback

````python
def ProxyResult(userInput:str,proxyInitial:ModelInitial,assistantInitial:ModelInitial,isReset:bool=True)->ProcessInitial:
  try:
    if (userInput is not None) and (userInput != " "):
      proxyInitial.initiate_chat(
          assistantInitial,
          message=str(userInput),
          llm_config=llmConfiguration,
          clear_history=False,
          handle_tool_error=True
      )
    for dct in assistantInitial.chat_messages.values():
      if dct is not None and len(dct) > 0:
        for dt in dct:
          if (dt["content"] != "TERMINATE") and (dt["content"] != str(userInput)) and (dt["content"] != "") and (dt["content"] is not None) and (dt["content"] != "CONTINUE"):
            content = content_str(dt["content"]).replace("TERMINATE","")
          else:
            pass
      else:
        content = "[The requested amount of data was not found]"
    if isReset:
      proxyInitial.reset()
      assistantInitial.reset()
    return content
  except:
    proxyInitial.reset()
    assistantInitial.reset()
    return "[The requested amount of data was not found]"
````
